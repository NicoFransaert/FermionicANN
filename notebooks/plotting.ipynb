```python
meta_file_H2 = '../data/RBM_runs/dissociation_H2/rbm_sto3g_H2_0-7348_eq1_a1_sgd_lr1_ns10000_trick.META'
hist_file_H2 = '../data/RBM_runs/dissociation_H2/rbm_sto3g_H2_0-7348_eq1_a1_sgd_lr1_ns10000_trick.log'
with open(meta_file_H2) as jf:
    meta = json.load(jf)
with open(hist_file_H2) as jf:
    hist = json.load(jf)
it = [i['Iteration'] for i in hist['Output']]
en = [i['Energy']['Mean'] + meta['SystemData']['nuc_rep_energy'] for i in hist['Output']]
exact = meta['SystemData']['tot_energy']

fig, ax = plt.subplots(1, 1, figsize=(5,4), squeeze=False, dpi=150)
ax = ax[0][0]
#plot energy history
# plt.plot(it, en)
# plt.title(r'$H_2$ @ equilibrium')
# plt.xlabel('Iteration')
# plt.ylabel('Energy')

#plot relative energy error
ax.plot(it, np.abs((np.array(en)-exact)/exact), color='b', label='H$_2$ RBM')
# plt.title(r'$H_2$ equilibrium')
ax.hlines(kcalmol, xmin=0, xmax=it[-1], linestyles='--', color='k', label='chemical accuracy')
ax.set_xlabel('iterations')
ax.set_ylabel(r'$\Delta E_0$ ($Ha$)')
ax.set_yscale('log')
ax.legend()
```

```python
meta_file = '../data/RBM_runs/H2_for_plot/rbm_sto3g_H2_0-7348_eq1_a1_sgd_lr01_ns10000_trick.META'
hist_file = '../data/RBM_runs/H2_for_plot/rbm_sto3g_H2_0-7348_eq1_a1_sgd_lr01_ns10000_trick.log'

with open(meta_file) as jf:
    meta = json.load(jf)
with open(hist_file) as jf:
    hist = json.load(jf)
it = [i['Iteration'] for i in hist['Output']][0:4000]
en = [i['Energy']['Mean'] + meta['SystemData']['nuc_rep_energy'] for i in hist['Output']][0:4000]
exact = meta['SystemData']['tot_energy']

fig, ax = plt.subplots(1, 2, figsize=(8,3.5), squeeze=False, dpi=150)
ax1 = ax[0][0]
ax2 = ax[0][1]

#plot relative energy error
ax1.plot(it, savgol_filter(np.abs((np.array(en)-exact)/exact),51,3), color='b')
# plt.title(r'$H_2$ equilibrium')
ax1.hlines(kcalmol, xmin=0, xmax=it[-1], linestyles='--', color='k', label='chemical accuracy')
ax1.set_xlabel('Iterations')
ax1.set_ylabel(r'$\Delta E_0$ ($Ha$)')
ax1.set_yscale('log')
ax1.legend()

#plot energy history
ax2.plot(np.array(it), savgol_filter(en,51,3), color='b')
ax2.hlines(-1.1170273947894147, xmin=0, xmax=it[-1], linestyles='dotted', color='k', label='HF energy')
# ax2.set_title('$H_2$ @ equilibrium')
ax2.set_xlabel('Iterations')
ax2.set_ylabel('Energy ($Ha$)')
ax2.set_ylim([-1.15,-1.0])
ax2.legend()

fig.tight_layout()
```

```python
bondlength = np.load('../data/dissociation/STO-3G/bondlength_H2.npy')
HF_energy = np.load('../data/dissociation/STO-3G/scf_energy_H2.npy')
FCI_energy = np.load('../data/dissociation/STO-3G/FCI_energy_H2.npy')
CCSD_energy = np.load('../data/dissociation/STO-3G/CCSD_energy_H2.npy')
CCSDT_energy = np.load('../data/dissociation/STO-3G/CCSD(T)_energy_H2.npy')

directory = '../data/RBM_runs/dissociation_H2/'
meta_files = []
for json_file in [f for f in os.listdir(directory) if f.endswith('.META')]:
    with open(directory + json_file) as jf:
        meta_files.append(json.load(jf))

rbm_bondlength = []
rbm_energies = []
rbm_sigma = []
for mf in meta_files:
    rbm_bondlength.append(mf['SystemData']['distance'])
    rbm_energies.append(mf['Total_energy']['Mean'])
    rbm_sigma.append(mf['Total_energy']['Sigma'])

fig, axs = plt.subplots(1, 1, figsize=(5,4), squeeze=False, dpi=150)

ax = axs[0][0]
ax.plot(bondlength, HF_energy, linestyle='--', c='blue', label='HF')
ax.plot(bondlength, FCI_energy, linestyle='--', c='red', label='FCI')
# ax.plot(bondlength, CCSD_energy, linestyle='--', linewidth=0.7, c='black', label='CCSD')
# ax.plot(bondlength, CCSDT_energy, linestyle='dotted', linewidth=0.7, c='black', label='CCSD(T)')
ax.errorbar(rbm_bondlength, rbm_energies, yerr=rbm_sigma, linestyle='', marker='x', c='green', capsize=3, label='RBM')
ax.set_xlabel('Nuclear separation ($\AA$)')
ax.set_ylabel('Energy ($Ha$)')
ax.legend(loc='upper right', fontsize='medium', ncol=2)

ins = ax.inset_axes([0.3,0.49,0.28,0.25])

ins.plot(bondlength, HF_energy, linestyle='--', c='blue')
ins.plot(bondlength, FCI_energy, linestyle='--', c='red')
# ins.plot(bondlength, CCSD_energy, linestyle='--', linewidth=0.7, c='black')
# ins.plot(bondlength, CCSDT_energy, linestyle='dotted', linewidth=0.7, c='black')
ins.errorbar(rbm_bondlength, rbm_energies, yerr=rbm_sigma, linestyle='', marker='x', c='green', capsize=3, label='RBM')
ins.set_xticks([0.65,0.734,0.8])
ins.set_xlim(0.65,0.8)
ins.set_ylim(-1.14, -1.13)
#plt.title('Probability')
ax.axis('tight')
fig.tight_layout()

# plt.savefig('../data/plots/STO-3G_H2_dissociation.png')
```
